```python
# -------------------------------------------------------------------
# 집계함수= => 평균, 표준편차, 최대값, 최소값, 중앙값, 최빈값,상관계수
# -------------------------------------------------------------------
import pandas as pd
```

```python
FILE = '../dataFiles/auto-mpg.csv'
mpgDF =pd.read_csv(FILE,header=None)
```

```python
# mpg0DF.info()
# type(mpgDF.tail(n=1)),type(mpgDF[0])

# mpgDF.iloc[-1], mpgDF.loc[0]

mpgDF.loc[:,[1,2]]

```

```python
# 컬럼 지정하기 => columns 속성

mpgDF.columns =['mpg', 'cylinders' ,'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'car_name']
mpgDF.head(3)
```

```python
# mpgDF.mean() #numeric_only=None' is deprecated.
# 숫자가 아닌 데이터들이 있다.
mpgDF.mean(numeric_only=True) #평균
```

```python
mpgDF.median(numeric_only=True) #중앙값
```

```python
mpgDF.mode() #최빈값 #가장 많이 나온 값
#동일한게 있다면 같이 뿌려줌
```

```python
mpgDF.horsepower.value_counts(),mpgDF.weight.value_counts(),
mpgDF.mpg.value_counts()
```

```python
print(mpgDF.max(), mpgDF.min(),sep="\n\n")
```

```python
#표준편차
mpgDF.std(numeric_only=True)
```

```python
#상관계수 => 컬럼과 컬럼의 연관성을 수치화 
# (깊은 상관관계)-1 ~ 0(관계없음) ~ 1 (깊은 상관관계)

mpgDF.corr(numeric_only=True)

```

```python
# auto_mpg 칼럼 중 서로 관련이 깊은 컬럼 순위 => 내림차순
oneCorr=mpgDF.corr(numeric_only=True).mpg
type(oneCorr),oneCorr.index
```

```python
#절대값 => 값기준 내림차순
oneCorr = abs(oneCorr)
oneCorr
```

```python
oneCorr.sort_values(ascending=False)
```

```python
# 데이터 프레임의 수치 데이터에 대한 기술통계값
# 평균, 표준편차, 최소값, 최대값
# mpgDF.describe().transpose()
newmpgDF = mpgDF.describe()

fourDF = newmpgDF.loc[['mean','std','min','max']]

fourDF = mpgDF.describe().iloc[[1,2,3,7]]

#행을 컬럼으로 , 컬럼을 행으로  => 전치(Transpose)

fourDF.transpose()
```

```python
fourDF.transpose()
```

```python
# mpg 컬럼의 데이가 25보다 큰 데이터만 출력하세요.
mpgDF.mpg > 25
# 이것을 활용해서, 시리즈형태의 결과값을 가지고 => 
mpgDF.mpg[5] #컬럼이 한 개 빠지면 씨리즈
mpgDF.mpg[mpgDF.mpg > 25]  # 비교연산을 거쳐서 나온, 데이터  => 특성 데이터로 추출할때, 

mpgDF.mpg[mpgDF.mpg > mpgDF.mpg.mean()]  #mpg의 평균보다 큰 애들 / 
# 불린 인덱싱! (Boolean Indexing) => True, False로 구성된 인덱스
# Ture인 행만 출력 
```

```python
mpgDF[mpgDF.mpg > mpgDF.mpg.mean()]
```
